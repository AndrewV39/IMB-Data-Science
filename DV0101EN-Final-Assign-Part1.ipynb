####  TASK 1.2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


<details><summary>Click here for a hint</summary>

<p>
    You will require:-
    <br>create a separate dataframe where the column recession has a value of '1'
    <br>to group the year, vehicle_type and calculate the average on the 'Automobile Sales'
    <br>one way is to -
    <br>use as_index as false else you will endup with multiple-indexed datafame
    <br>later set year as index and groupby vehicle over Sales and plot
    <br>make use of .plot() with kind = 'line'
    <br>do not forget to include labels and title
</p>
</details>


````python
import matplotlib.pyplot as plt

# Step 1: Group by Year and Vehicle_Type, then get mean sales
grouped = df.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Step 2: Pivot to have Vehicle_Type as separate columns
pivot_df = grouped.pivot(index='Year', columns='Vehicle_Type', values='Automobile_Sales')

# Step 3: Plot each Vehicle_Type line
plt.figure(figsize=(12, 7))
pivot_df.plot(kind='line', marker='o', ax=plt.gca())

# Step 4: Recession Annotations
if 1982 in pivot_df.index:
    plt.axvline(x=1982, color='gray', linestyle='--', alpha=0.7)
    plt.text(1982, plt.ylim()[1]*0.95, '1981–82 Recession', rotation=90, color='gray')
if 2009 in pivot_df.index:
    plt.axvline(x=2009, color='gray', linestyle='--', alpha=0.7)
    plt.text(2009, plt.ylim()[1]*0.95, '2008–09 Recession', rotation=90, color='gray')

# Step 5: Labels, Title, and Legend
plt.xlabel('Year')
plt.ylabel('Average Sales')
plt.title('Automobile Sales Trends by Vehicle Type (Highlighting Recessions)')
plt.xticks(list(range(1980, 2024)), rotation=75)
plt.legend(title='Vehicle Type')
plt.tight_layout()
plt.grid(True)
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('historical_automobile_sales.csv')

# Filter and sort data
rec_data = df[df['Recession'] == 1].sort_values('Year')
non_rec_data = df[df['Recession'] == 0].sort_values('Year')

# Create figure and subplots
fig = plt.figure(figsize=(12, 6))

# Recession subplot
ax0 = fig.add_subplot(1, 2, 1)
sns.lineplot(x='Year', y='GDP', data=rec_data, ax=ax0, errorbar=None)
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')
ax0.set_title('GDP Variation During Recession')
ax0.set_ylim(20, 70)

# Non-Recession subplot
ax1 = fig.add_subplot(1, 2, 2)
sns.lineplot(x='Year', y='GDP', data=non_rec_data, ax=ax1, errorbar=None)
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.set_title('GDP Variation During Non-Recession')
ax1.set_ylim(20, 70)

# Final layout
plt.tight_layout()
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter non-recession data
non_rec_data = df[df['Recession'] == 0]

# Create Bubble Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=non_rec_data,
    x='Month',
    y='Automobile_Sales',
    size='Seasonality_Weight',
    hue='Seasonality_Weight',
    sizes=(20, 300),   # controls bubble size range
    legend=False       # remove legend if not needed
)

# Labeling
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Seasonality Impact on Automobile Sales')

plt.tight_layout()
plt.show()
````

 ### TASK 1.9: Develop a lineplot to analyse the effect of the unemployment rate on vehicle type and sales during the Recession Period.
 <br>Analyze the effect of the unemployment rate on vehicle type and sales during the Recession Period
 #### You can create a lineplot and title the plot as 'Effect of Unemployment Rate on Vehicle Type and Sales'


<details><summary>Click here for a hint</summary>

<p>
    Filter out the data for recession period<br>
     Make use of lineplot() from seaborn and pass the relavent data</p>
</details>


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter and clean data for Recession period
df_rec = df[df['Recession'] == 1].copy()

# Strip whitespace, replace spaces, and lowercase column names
df_rec.columns = df_rec.columns.str.strip().str.replace(" ", "_").str.lower()

# Drop rows with missing values in required columns
df_rec = df_rec.dropna(subset=['unemployment_rate', 'automobile_sales', 'vehicle_type'])

# Plot lineplot
sns.lineplot(
    data=df_rec,
    x='unemployment_rate',
    y='automobile_sales',
    hue='vehicle_type',
    style='vehicle_type',
    markers='o',
    dashes=False,
    err_style=None
)

# Formatting
plt.ylim(0, 850)
plt.xlabel('Unemployment Rate')
plt.ylabel('Automobile Sales')
plt.title('Effect of Unemployment Rate on Vehicle Type and Sales')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
````
